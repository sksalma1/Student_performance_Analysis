```python
dim_student = stdInfo[[
    'id_student', 'gender', 'region', 'highest_education',
    'imd_band', 'age_band', 'num_of_prev_attempts',
    'studied_credits', 'disability', 'final_result'
]].drop_duplicates()

dim_student.head()
```

```python
dim_course = courses.rename(columns={
    'code_module': 'course_module',
    'code_presentation': 'course_presentation',
    'module_presentation_length': 'presentation_length'
})

dim_course.head()
```

```python
dim_assessment = ass.rename(columns={
    'code_module': 'course_module',
    'code_presentation': 'course_presentation'
})

dim_assessment.head()
```

```python
dim_vle = vle[['id_site', 'code_module', 'code_presentation', 'activity_type', 'week_from', 'week_to']].drop_duplicates()
dim_vle.head()
```

```python
dim_registration = stdReg[['id_student', 'code_module', 'code_presentation', 'date_registration', 'date_unregistration']].drop_duplicates()
dim_registration.head()
```

```python
fact_student_performance = stdAss.merge(
    ass[['id_assessment', 'code_module', 'code_presentation', 'weight']],
    on='id_assessment',
    how='left'
).merge(
    stdvle.groupby(['id_student'])['sum_click'].sum().reset_index().rename(columns={'sum_click': 'total_clicks'}),
    on='id_student',
    how='left'
)

fact_student_performance.head()
```

```python
from sqlalchemy import create_engine
engine = create_engine("postgresql+psycopg2://postgres:Salma123%40@localhost:5432/Education_DWDM")
```

```python
# Sending each DataFrame to PostgreSQL
dim_student.to_sql('dim_student', engine, if_exists='replace', index=False)
dim_course.to_sql('dim_course', engine, if_exists='replace', index=False)
dim_assessment.to_sql('dim_assessment', engine, if_exists='replace', index=False)
dim_vle.to_sql('dim_vle', engine, if_exists='replace', index=False)
dim_registration.to_sql('dim_registration', engine, if_exists='replace', index=False)
fact_student_performance.to_sql('fact_student_performance', engine, if_exists='replace', index=False)

print("All tables uploaded to PostgreSQL successfully!")
```

```python
import matplotlib.pyplot as plt
df = pd.read_sql("SELECT code_module, AVG(score) as avg_score FROM fact_student_performance GROUP BY code_module;", engine)

# Plot
plt.figure(figsize=(10,5))
plt.bar(df['code_module'], df['avg_score'], color='skyblue')
plt.xlabel('Course Module')
plt.ylabel('Average Score')
plt.title('Average Score per Course Module')
plt.show()
```
